**Artificial Narrow Intelligence (ANI)**: 
- Specialized in one task, like facial recognition or language translation. Current real-world AI falls into this category. 
- Designed for a specific task and cannot operate outside its trained scope. 
- Examples: 
    - ChatGPT (text-based conversation), Siri, Alexa, facial recognition systems. 
    - Robotics Use: Industrial robots performing repetitive tasks or autonomous drones with fixed functions.

**Artificial General Intelligence (AGI)**: 
- Hypothetical AI that can perform any intellectual task a human can. 
- Theoretical at present. 
- Would perform any intellectual task a human can, 
- adapting knowledge across domains without retraining. 
- Could enable robots to learn new skills in real-time.

**Artificial Super Intelligence (ASI)**: 
- Hypothetical AI that surpasses human intelligence. 
- in reasoning, creativity, and emotional understanding. 
- Would have independent goals, beliefs, and possibly emotions.
