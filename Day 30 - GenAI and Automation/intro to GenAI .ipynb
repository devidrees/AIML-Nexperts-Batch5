### **Timeline**

Year | Model | Description
---|---|---
2014 | GANs (Generative Adversarial Networks) | first major architecture capable of generating realistic images
2016 | WaveNet by Google DeepMind | realistic AI speech synthesis system. Robotic TTS (Text to Speech) systems replaced with human like audio | Autogregressive Networks predicted audio waveform using CNNs trained on speech data
2017 | Transformer Architecture | Foundation of modern GenAI. Enabled models to understand long range context in text. Used the Self-Attention Mechanism
2018 | GPT-1 First large GPT | Trained on large text corpus
2019 | GPT-2 | 1.5B parameters. Worked on Auto regressive generation: predict the next token repeatedly
2020 | GPT-3 | major breakthrough - 175B parameters. Near human level writing and few shot learning
2021 | DALL.E, Codex (code generation)| OpenAI text to image generative model. Learned mapping text to image representations
2022 | Stable Diffusion | image generation, thus AI art became possible
2023 | GPT-4 (Multimodal) | Text + image. imporoved alighment via RLHF (Reinforecment Learning from Human Feedback)
2023 | Claude 2/3 | By Anthropic. Focused on sagety and long context reasoning. 
2023 | Gemini | Deepmind, native multimodal generative model - text, image, audio and code
2024 | SORA (OpenAI), Runway Gen-3 (Runway) | Text to Video model. Major leap in Video generation. Diffusion-Transformer
2025 | Advanced Long-Context LLMs (`1M+ tokens`) | Models capable of processing entire books, codebases, and long documents. 
2025 | Real Time Multimodal Agents | AI systems that generate text, speech, visuals simultaneously in `live interaction`. Multimodal Transformers + streaming generation + memory modules for real time responses
2026 | Autonomous Generative Agents | capable of planning, generating, revising, executing multistep creative workflows | Combine LLM reasoning, tool usage, memory persistance, retrival systems (RAG), self reflection loops for iterative content creation

How to get your Gemini API Key

1. Go to Google AI Studio: Visit aistudio.google.com.

2. Sign In: Use your standard Google/Gmail account.

3. Create the Key: On the left-hand sidebar, click the "Get API key" button.

4. Generate: Click "Create API key" (you can choose to create it in a new project).

5. Copy: A long string of letters and numbers will appear. Copy this immediately and keep it secret!

Replace "your_api_key_here" with the long string you just copied. 

>  Pro-Tip: If you plan on sharing your code or putting it on GitHub, never hardcode the key directly in the file. 
- Instead, set it as an environment variable on your computer so the script can "find" it without it being written in plain text
