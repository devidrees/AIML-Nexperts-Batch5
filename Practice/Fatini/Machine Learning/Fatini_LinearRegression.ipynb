```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 100)

# Define the equation
y = 10 * x + 5
y2 = 1 / (1 + np.exp(-x))

fig, ax = plt.subplots(1, 2, figsize = (12,4))

ax[0].plot(x,y)
ax[0].axhline(0, color = 'red', linewidth = 0.5)
ax[0].axvline(0, color = 'red', linewidth = 0.5)
ax[0].set_title("Linear Function")
ax[0].grid()

ax[1].plot(x,y2)
ax[1].axhline(0, color = 'red', linewidth = 0.5)
ax[1].axvline(0, color = 'red', linewidth = 0.5)
ax[1].set_title("Sigmoid Curve")
ax[1].grid()

# Show the graph

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 100)

# Sigmoid function
y = 1 / (1 + np.exp(-x))

# Plot
plt.plot(x, y)

# Labels and title
plt.xlabel("x")
plt.ylabel("y")

# Show the graph
plt.show()
```
