```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
transactions = [
    ['Milk', 'Butter', 'Bread'],
    ['Milk', 'Bread'],
    ['Bread', 'Butter', 'Eggs'],
    ['Milk', 'Eggs'],
    ['Bread', 'Butter', 'Juice']    
]
```

```python
te = TransactionEncoder()
te_array = te.fit_transform(transactions) #.transform(transactions)

te.columns_

df_transactions = pd.DataFrame(te_array, columns=te.columns_)
df_transactions
```

```python
frequent_itemsets = apriori(df_transactions,
                         min_support = 0.4,
                         use_colnames = True)
```

```python
rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=0.4)

final_table = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

display(final_table)
```

```python
current_cart = frozenset({'Milk'})

recommendations = final_table[final_table['antecedents'] == current_cart]

recommendations
```
