```python
import numpy as np

# ======================================
# Scenario 1
# ======================================

# 1. Setup
np.random.seed(42)

# 2. Generate scores for 10 students and 3 subjects
scores = np.random.randint(0, 101, size=(10, 3))

print("Scores Array:")
print(scores)

# 3. Array Inspection
print("\nArray Information")
print("Dimensions (ndim):", scores.ndim)
print("Shape:", scores.shape)
print("Total Size:", scores.size)

# 4. Student IDs (100 to 109)
student_ids = np.arange(100, 110)

print("\nStudent IDs:")
print(student_ids)

# 5. Average score for each subject
subject_average = np.mean(scores, axis=0)

print("\nAverage Score Per Subject:")
print(subject_average)

# 6. Boolean Masking (scores below 50)
failing_scores = scores[scores < 50]

print("\nFailing Scores (<50):")
print(failing_scores)

# 7. Automated Grading
report_card = np.where(scores < 50, "Fail", "Pass")

print("\nReport Card:")
print(report_card)
```

```python
# ======================================
# Scenario 2
# ======================================

# 1. Numerical Data
student_ids = np.arange(101, 106)

# Example final scores for 5 students
scores = np.array([78, 45, 92, 50, 33])

# 2. Create Pass/Fail Status
status = np.where(scores >= 50, "Pass", "Fail")

# 3. Merge using column_stack
final_report = np.column_stack((student_ids, scores, status))

print("\nFinal Report:")
print(final_report)

# 4. Check data type
print("\nData Type of Final Report:")
print(final_report.dtype)

# Try adding bonus points
try:
    final_report[:, 1] = final_report[:, 1].astype(int) + 5
    print("\nAfter Adding Bonus Points:")
    print(final_report)
except Exception as e:
    print("\nError:")
    print(e)
```
