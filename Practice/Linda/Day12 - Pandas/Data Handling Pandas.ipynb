# Selection using the location or labels
    - Location -- .iloc[]
    - df.iloc[row index] #access  :stepsize

    - df.iloc[row_index_slice] # start:end
    - df.iloc[row_slice,column]


```python
df.iloc[0] #first row as a series

df.iloc[1:4] #rows from 1 up to 3

df.iloc[1:4:2] #in the step 2

df.iloc[-1] # last row
# display(df)

# --- indexing work as well
df.iloc[:4,1:3] #first 4 rows

# df.iloc[[0,2,3],[0,2,4]] # you can access the rows, columns

```

```python
# i want the data of the last two entries for department & experince

print(df.iloc[-2:,[1,3]])
```
