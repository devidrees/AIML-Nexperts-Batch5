```python
Core Objects

- Series - is a one dimensional labelled array (like a single column)
    - series = one column
- DataFrame - a two-dimensional table made of multiple series of same index
    - dataframe = a sheet of columns and rows
```

```python
pd.Series([20,30,40])
```

```python
marks = pd.Series([20,30,40])
type(marks) #pandas.Series
print(marks)
```

```python
maths = pd.Series([20,30,40], dtype = 'int8', name = 'maths')
type(maths) #pandas.Series

print(maths)
```

```python
english = pd.Series([20,30,40])
python = pd.Series([20,55,40])
science = pd.Series([20,99,40])
```

```python
import pandas as pd

pd.concat([maths, english, python, science]) # # series objects

df =pd.concat([maths, english, python, science], axis = 1) # dataframe

print(type(df)) #pandas.DataFrame
print(df)
```

```python
# dataFrame from dictionary of Lists --- each list need to be in the same size
data = {
    "Students": ['Nomie', 'Idrees', 'Radia'],
    "Maths": [85,95,90],
    "English": [75,85,80],
}
df = pd.DataFrame(data) # use this to create a dataframe from a dictionary
df # final for recall results of coding
# print(df)   
# display(df)
```

```python
# dataFrame from list of dictionary --- each dictionary can be of diff size
# we can miss elements ---- those elements will appear as Nan values
data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'London'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Paris'}
]
 
df = pd.DataFrame(data)
# print(df)
df
```
