```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

df.head()
```

```python
model_kmeans = KMeans(n_clusters=3, init = "random",random_state=42)

model_kmeans

#model.fit(df)
```

```python
df["kClustering"] = model_kmeans.fit_predict(df)
```

```python
# how far spread out the clusters are from each other
# how good the clusters are formed
model_kmeans.inertia_
```

```python
df.sample(5)
```

## Visualise

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=df["kClustering"])
ax[0].set_title("Clustering by K Means")

ax[1].scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=data.target)
ax[1].set_title("Clustering using the species column")

plt.tight_layout()
plt.show()

#plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=df["kClustering"])

#plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"], c=data.target)

#plt.show()
```

```python
# Elbow Method

error = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    error.append(kmeans.inertia_)

```

```python
#errors
plt.plot(range(1, 11), error, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Error')
plt.title('Elbow Method')
plt.show()
```
