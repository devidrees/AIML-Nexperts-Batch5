Diff types of EDA in Data Visualisation part:
1. Univariate - analysis of only one variable
2. Bivariate - of 2 variables
3. Multivariate - of more than 2 variables
- does not only mean for visualisation

EDA can be:
1. Quantitative - descriptive - summary tables, central tendencies (mean, median, mode), dispersion, variance
2. Graphical - Visual - Bar Chart, histogram, scatter plot, box plot etc

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = [1,2,3,4,5]
values = [100,20,100,40,80]

bar_color = ["red", "pink", "orange", "yellow", "green"]

# hashvalues are used for colors in web
# hexa decimal is a number system - 0 1 2 3 4 5 6 7 8 0 A B C D E F
bar_color1 = ["#fc5757", "#b83535", "#460808", "#1B0101", "#3d0808"]
#bar_color1 = #ffffff

hatch = ['\\\\', '|||', '+','O', 'X']

#bar chart
plt.bar(categories, values, 
        color = bar_color1, # color argument decides the colors at bars
        edgecolor = "black",   #"black" or #ffffff
        linewidth = 2,
        alpha = 0.7, # transperancy
        hatch = "\\"  #"\\"
        ) 

# title
plt.title("Bar Chart For Show Only")

# label for x axis
plt.xlabel("categories")

# label for y axis
plt.ylabel("values")

plt.show()
```

```python
# Exercise:
    ## build a scatterplot of total_bill vs tip in the tips dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Load tips dataset
#tips = sns.load_dataset("tips")
df_tips = sns.load_dataset("tips")

sns.scatterplot(
    data=df_tips,
    x="total_bill",
    y="tip",
    s = 50,
    hue = "day"
)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill (RM)")
plt.ylabel("Tip (RM)")

plt.show()
```

A box plot, also known as a box-and-whisker plot, is a graphical representation
used in descriptive statistics to display the distribution of a dataset
based on a five-number summary:
minimum,
first quartile (Q1),
median,
third quartile (Q3), and
maximum

```python
# Compute the correlation matrix
# Remove the 'species' column because it's non-numeric, then calculate correlation on numeric features
corr_matrix = df_iris[['sepal_length',	'sepal_width',	'petal_length',	'petal_width']].corr()

corr_matrix = df_iris.drop('species', axis = 1).corr().round(2)
corr_matrix
# which two features have the strongest relationship???
# 
```

```python
# Create a heatmap
plt.figure(figsize=(8, 6))  # Set the size of the plot (width=8, height=6)

# Create heatmap using seaborn's heatmap() function
sns.heatmap(
    corr_matrix,       # The correlation matrix
    annot=True,        # Annotate each cell with the numeric value
    cmap='inferno',   # Color map (coolwarm ranges from blue for low to red for high correlation)
    # try these themes for cmap: Blues, BuGn, BuPu, magma, inferno,plasma
    linewidths=0.5,    # Width of the lines that will divide each cell
    linecolor='black', # Color of the lines dividing cells
    vmin=-1,           # Minimum color scale value (for -1 correlation)
    vmax=1,            # Maximum color scale value (for 1 correlation)
    cbar_kws={"shrink": 0.8}  # Customize color bar (shrink it slightly)
)
```

```python
plt.pie((titanic['class'].value_counts()), 
        # labels= ['Not Survived','That Lived'], 
        labels = ['Third', 'Second', 'First'],
        autopct='%.2f%%', 
        startangle=90, 
        colors = sns.color_palette('dark'))

plt.show()
```

```python
# Donut Chart


# colors =  ['red','green'] 
labels= ['Drowned','That Lived']
colors = sns.color_palette("pastel")[0:2]
percentage='%.1f%%'
wedgeprops={"width": 0.6} # donut chart

plt.pie(titanic['survived'].value_counts(), 
        labels=labels, 
        colors = colors, 
        autopct=percentage, 
        startangle=90, 
        wedgeprops= wedgeprops)

plt.show()
```

```python
titanic = sns.load_dataset("titanic")

# Create pie chart
plt.pie(
    titanic["class"].value_counts(),
    labels=["Third", "First", "Second"],
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("dark")
)

plt.title("Titanic Passenger Class Distribution")

plt.show()
```

```python
# Exercie
## Smoker & non smoker

#import seaborn as sns
#import matplotlib.pyplot as plt

# Load tips dataset
tips = sns.load_dataset("tips")

# Count smoker and non-smoker
smoker_count = tips["smoker"].value_counts()

plt.pie(smoker_count,
        labels=smoker_count.index,
        autopct="%1.1f%%",
        startangle=90,
        colors=sns.color_palette("dark"))

plt.title("Smoker vs Non-Smoker Distribution")

plt.show()
```
