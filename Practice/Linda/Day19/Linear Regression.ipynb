```python
## 1. 

import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 100)

# Define equation y = 10x + 5
y = 10 * x + 5

# Plot
plt.plot(x, y, label="y = 10x + 5")
plt.title("Linear Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
## 2. 

import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 100)

# Define sigmoid function
y = 1 / (1 + np.exp(-x))

# Plot
plt.plot(x, y, label="Sigmoid Function")
plt.title("Sigmoid (Logistic) Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt

x = [-10, 0, 10]

y = [10*i + 5 for i in x]

plt.plot(x, y, marker="o")
plt.title("Linear Function")
plt.grid(True)
plt.show()
```
