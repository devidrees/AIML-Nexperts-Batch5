```python
import seaborn as sns

data = pd.Series([20,20,20,25,25,30,100,30,30,35,40,45,50,70])

print(data.median())

#sns.histplot(data)

#sns.histplot(data, bins =10 )

#sns.boxplot(data)

data.describe() #quantile

data.quantile() # min
Q1 = data.quantile(0.25) #1st quartile Q1
data.quantile(0.50) #2nd quartile
Q3 = data.quantile(0.75) #3rd quartile Q#
data.quantile(1) # max

IQR = Q3 - Q1
# IQR method of outlier detection
lower = IQR - 1.5 * IQR
upper = IQR + 1.5 * IQR

print(lower,upper)

# filter out the outliers
outlier = data[(data < lower) | (data > upper)]
outlier
```

Inter Quartile Range:
- IQR = Q3 - Q1

- to define normal data:
    - we decide what the lower bound & upper should be
    - anything beyond is an outliner

- lower = IQR - 1.5 * IQR (1.5 is a standard number)
- upper = IQR + 1.5 * IQR
