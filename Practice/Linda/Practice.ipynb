The Scenario 1:
 
- The school's database has just exported a batch of raw exam scores. You need to use NumPy to generate some sample data, inspect its structure, and automatically grade the students based on a curve.
 
Your Tasks:
 
1. The Setup: Import NumPy. To ensure the whole class gets the exact same "random" numbers, set the random seed to 42.
 
2. Data Generation: Use random.randint to generate a 2D array of random scores between 0 and 100. The array should represent 10 students and their scores across 3 different subjects.
 
3. Array Inspection: Prove you built the array correctly by printing its ndim (number of dimensions), its shape, and its total size (total number of scores).
 
4. Student IDs: Use arange to create a 1D array of Student IDs starting at 100 and ending at 110.
 
5. Perform the aggregate operation of average and find the average score of students for each subject
 
6. The Audit (Boolean Masking): The principal wants to know every instance where a student scored below a 50. Use boolean masking to extract and print just those failing scores from the main array.
 
7. Automated Grading (where): Use np.where to create a brand new "Report Card" array. If a score in the original array is less than 50, it should say "Fail". If it is 50 or above, it should say "Pass".

```python
# Task 1: Import NumPy and set random seed
print(np.random.seed(42))

# Task 2: Generate a 2D array of random scores
# 10 students, 3 subjects, scores between 0 and 100
scores = np.random.randint(0, 101, (10, 3))

print("Exam Scores: ")
print(scores)

# Task 3: Array Inspection
print("\nArray Inspection")
print("Number of Dimensions (ndim):", scores.ndim)
print("Shape:", scores.shape)
print("Total Size:", scores.size)

# Task 4: Create Student IDs from 100 to 109
student_ids = np.arange(100, 110)

print("\nStudent IDs:")
print(student_ids)

```

```python
# Task 5: Average score for each subject
subject_average = np.mean(scores, axis=0)

print("\nAverage Score for Each Subject:")
print(subject_average)

# Task 6: Boolean Masking
# Extract scores below 50

failing_scores = scores[scores < 50]

print("\nFailing Scores (Below 50):")
print(failing_scores)

# Task 7: Automated Grading using np.where

report_card = np.where(scores < 50, "Fail", "Pass")

print("\nReport Card:")
print(report_card)


```

Scenario 2:
 
The semester is over. The principal wants a final report that includes Student IDs, their final scores, and a text label indicating whether they Passed or Failed. You need to combine all this data into one final grid and export it so it can be sent to the administration.
 
Your Tasks:
 
1. The Numerical Data: Access the 1D marks arrays of 5 Student IDs (101 to 105), store separately in variables .
 
2. The Text Data: Use `np.where` to create a status array. If the score is 50 or above, the status is "Pass". If it's below 50, it is "Fail".
 
3. The Merge (column_stack): The principal wants this data side-by-side. Use np.column_stack to merge the IDs, the Scores, and the Statuses into one single 2D array called final_report. Print it to the screen.
 
4. Data Types: The administration suddenly wants you to add 5 bonus points to the scores in your final_report. Try to add 5 to the score column. What happens? Print the .dtype (data type) of your final_report to figure out why.
 
(OPTIONAL)
 
5. The Export: Even though the math broke, save your final_report to your computer as a CSV file called "edupi_final_grades.csv" so another program can deal with it.
 

```python
# Task 1:

import numpy as np

# Student IDs
student_ids = np.arange(101, 106)

# Final Scores
scores = np.array([85, 42, 67, 49, 91])

print("Student IDs:")
print(student_ids)

print("\nScores:")
print(scores)

# Task 2

status = np.where(scores >= 50, "Pass", "Fail")

print("Status:")
print(status)


# Task 3

final_report = np.column_stack((student_ids, scores, status))

print("Final Report:")
print(final_report)
```

```python
# task 4

score_column = final_report[:, 1].astype(int)

bonus_scores = score_column + 5

print(bonus_scores)


#task 5

np.savetxt(
    "edupi_final_grades.csv",
    final_report,
    delimiter=",",
    fmt="%s",
    header="StudentID,Score,Status",
    comments=""
)

print("CSV file saved successfully!")
```
