## Arrays
- Arrays is a homogenous and linear data structure
    - homogeous - you save only one type of elements inside
    - it has indicies - we can use them to access data
    - Contigous blocks of memory - every eloement comes one after the other 

- if we already have list in python, why do we need arrays?
    - Array are very `FAST` compared to list
        - reason is not just how numpy no

List | Array
-----|------
Store pointer(address of the data)  |  Store the actual data
object are scattered through the memory  |  are Contigous (one after the other)
CPU has to go to each object address to access data  |  CPU accesses base address and eg. 8 to that address to get the next element
