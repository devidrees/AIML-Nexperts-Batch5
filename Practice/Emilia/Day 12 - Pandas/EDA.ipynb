### EDA (Exploratory Data Analysis) with Pandas

- first step in data analysis and ML too
- three parts:
    1. Data Cleaning - identify and remove missing values, or outliers etc
    2. Data Summarization - check the dataset characteristics and descriptive statistics
    3. Data Visualization - visualize complex data into visuals for easy interpretation 

- Some methods that we can run on dataframes
- Basic Inspection methods: (Data Summarization)
    - `df.head()` - give you the first 5 elements
        - an argument of 10 will give you first 10 elements
    - df.tail() - last 5 but argument can change the number
    - `df.sample()` - one random entry from the dataset
        - an argument can change that number too 
        - we use this to check how our datasets look like inside
- Basic attributes of a dataframe:
    - `df.shape` - tells you the shape of the df --- how many elements along how many axes
    - df.size - how many elements there are
    - df.ndim - will tell you the number of axes there are in the df --- generally just 2
    - `df.dtypes` - will tell you the data type of each column



- Methods for Statistical information as well:
    - `df.info()` - general information about the df
        - how many records, how many columns, what is the data type of those columns
        - also gives an idea of if there are any missing values in data
    - `df.describe()` - gives us the statistical measurements of the columns like count, mean, min, max, std, percentiles

## Aggregation
- Sometimes we need to get aggregate values of columns
- methods: sum(), min(), max(), mean()
- 
- AXIS --- 0 >>>>> Column Wise
- AXIS --- 1 >>>>> Row Wise

Use of axis argument 3 places:
- `Aggregate` --- Numpy 2D array --- DataFrame
    - axis = 0 ---> Column wise aggregates
    - axis = 1 ---> Row wise aggregates
    
- `Drop` a column or row in dataframes
    - axis = 0 ---> # Drop ROW
    - axis = 1 ---> # Drop COLUMN

```python
df_emp = pd.read_csv('employee.csv')
df_emp
```

```python
df_emp[['Salary','Experience']].mean(axis = 0) # axis = 0  Column wise

df_emp[['Salary','Experience']].mean(axis = 1) # ROW WISE aggregates
```

```python
# data cleaning
# drop dropna

# targetting the first ROW header, and finding the Column name there
df_emp.drop('Name', axis = 1) #  Drop COLUMN
df_emp.drop(0, axis = 0) # Drop ROW
```

### Combine DataFrames
- two dataframe and you want to combine them together
- 

JOINS - 
- are used to combine two tables
- this happens along a KEY column ---> PRIMARY KEY --- we cover this in SQL
- 
- ![](https://tse4.mm.bing.net/th/id/OIP.lRkZ1jamrvFYbSdGu3nF5gAAAA?w=350&h=350&rs=1&pid=ImgDetMain&o=7&rm=3)

```python
# Student info
students = {
    "StudentID": [1, 2, 3],
    "Student": ["Aman", "Sara", "John"]
}
df_students = pd.DataFrame(students)
display(df_students)

# Marks info
marks = {
    "StudentID": [2, 3, 4],
    "Math": [85, 90, 78],
    "English": [88, 92, 80]
}
df_marks = pd.DataFrame(marks)
display(df_marks)
```

```python
# Merge
# INNER JOIN 
df_merged_IJ = pd.merge(df_students, df_marks, on = "StudentID", how = 'inner')
# `on` will be a column that is inside both tables, and each row in this column, should be unique 

# All ID elements should match

df_merged_IJ
```

```python
# LEFT JOIN
df_merged_LJ = pd.merge(df_students, df_marks, on = "StudentID", how = 'left')
df_merged_LJ
```

```python
# RIGHT JOIN
df_merged_RJ = pd.merge(df_students, df_marks, on = "StudentID", how = 'right')
df_merged_RJ
```

```python
# OUTER JOIN
df_merged_FJ = pd.merge(df_students, df_marks, on = "StudentID", how = 'outer')
df_merged_FJ
```
