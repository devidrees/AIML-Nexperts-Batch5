super().__init__() # is used to call a parent class

    * Parent=====animal

````python
# Parent class
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self._age = age 

    def introduce(self):
        print(f"Hi. I am {self.name}!!\nI am a {self.species}\nI am {self._age} years old!")

    # Getter Method
    def get_age(self):
        return self._age
    
    # Setter Method
    def set_age(self, new_age):
        self._age = new_age
        return f"The new age is {new_age}"
    
    def have_birthday(self):
        self._age +=1
        print(f"It's {self.name}'s birthday. Now {self._age} years old ")
````

````python
# CHILD of ANIMAL
class Dog(Animal):
    def __init__(self, name, age): # thos is what we will pass ourselves
        # super() refers to the parent class --- could have used Animal as well
        #
        super().__init__(name, "Dog", age)
    
    def sound(self):
        print(f"{self.name} says: Woof!!")

    def detect_drug(self):
        if self.age > 5:
            return f"I will sniff out any drug"
        else:
            return f"{self.name} says, I am just {self.age} years old. You fr??"
````

````python
# CHILD of ANIMAL
class Cat(Animal):
    def __init__(self, name, age):
        # super() refers to the parent class --- could have used Animal as well
        #
        super().__init__(name, "Cat", age)
    
    def sound(self):
        print(f"{self.name} says: Meow!!")
````
