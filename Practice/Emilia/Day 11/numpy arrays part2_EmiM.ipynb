### Boolean Masking - (Filtering)  
- used to filter elements out

- array[mask] === mask is a condistion
- array_name[<condition>]


```python
scores = np.array([45, 71, 88, 30, 55, 90])

# broadcasting works on arrays
scores > 50 # array([False,  True,  True, False,  True,  True])
# element wise comparison
```

```python
# Boolen masking
scores[scores > 50] # whwreever this condition is true
```

```python
# where
age = np.array([5,17,24, 51, 23, 14, 8, 29, 33, 44, 10])

voting_eligiblity = np.where(age >= 18, "Eligible", "Not-Eligible")
print(age)
print(voting_eligiblity)
print("---"*5)

zip_list = list(zip(age, voting_eligiblity))
# zip - packs two iterables of equal length together

for age, eligibility in zip(age, voting_eligiblity):
    print(f"{age} : {eligibility}")
```
