```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
 
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# bring the data
 
iris = load_iris()
 
X, y = iris.data, iris.target
 
X_train, X_test, y_train, y_test = train_test_split(X, y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
y
```

```python
# Model

model = keras.Sequential([
    # input layer : no neurons, just a placeholder for innput
    layers.Input(shape = (4,)),

    # hidden layer : 
    layers.Dense(units = 5, activation = "relu"),

    #output layer : 
    layers.Dense(units = 3, activation = "softmax")
])


# model compilation - compile the model
model.compile(
    optimizer = 'adam', # optimzer in gradient descent which helps model to redduce the error to fnd least possible error - Global Minima
    loss = 'sparse_categorical_crossentropy', # function used to calculate the loss, most common sparse_categorical_crossentropy (multiclass & integer label) & categorical crossentropy
    metrics = ['accuracy']
)
```

```python
# Training

model.fit(X_train, y_train,
epochs = 100,
batch_size = 10,
validation_split = 0.3
)
```

```python
# Evaluation

predictions = model.predict(X_test)
predictions.round(2)

# confidence of each cateogry
# prediction is made by taking highest value

```

```python
y_pred = np.argmax(predictions, axis = 1) # the index of the highest value from each row
y_pred
```

```python
# to see the outputs
for i in range(len(predictions)): 
    print(f"Predicted Class: {np.argmax(predictions[i])} | Actual: {y_test[i]}")
```
