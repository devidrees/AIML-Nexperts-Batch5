```python
# 2A. Data Visualization
# Visualization 1: Revenue Trend

sales_year = df.groupby("Year")["Order_Amount"].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x=sales_year.index,y=sales_year.values,marker="o")
plt.title("Revenue Trend", fontsize=16, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Revenue (mil)")

# Display year labels explicitly
plt.xticks(ticks=[2023, 2024, 2025, 2026],labels=["2023", "2024", "2025", "2026"])

# Add revenue value labels on points
for x, y in zip(sales_year.index, sales_year.values):
    plt.text(x,y,f"${y/1e6:.2f}M",ha="center",va="bottom",fontsize=10)
plt.grid(alpha=0.3)

plt.show()
```

```python
# Visualization 3: Country vs Gender Analysis

# Country vs Gender count
country_gender = pd.crosstab(df["Country"],df["Customer_Gender"])

# Convert to percentage
country_gender_pct = (country_gender.div(country_gender.sum(axis=1), axis=0)* 100)

# Create stacked bar using percentage data
ax = country_gender_pct.plot(kind="bar",stacked=True,figsize=(12,6))

plt.title("Customer Gender Distribution Across Countries (%)",fontsize=16,fontweight="bold")

plt.xlabel("Country")
plt.ylabel("Percentage (%)")

# Add percentage labels at the center of each color segment
for i, country in enumerate(country_gender_pct.index):

    bottom = 0

    for gender in country_gender_pct.columns:

        value = country_gender_pct.loc[country, gender]

        if value > 3:
            ax.text(i,bottom + value/2,f"{value:.1f}%",ha="center",va="center",fontsize=10,fontweight="bold",color="black")

        bottom += value
plt.xticks(rotation=45)
plt.legend(title="Gender",bbox_to_anchor=(1.05,1),loc="upper left")
plt.tight_layout()

plt.show()
```

# 3. Data Cleaning Pipeline

1. Explore Dataset
        ↓
2. Detect Missing Values
        ↓
3. Remove Duplicate Records
        ↓
4. Fix Outliers
        ↓
5. Correct Data Errors
        ↓
6. Fix Data Types
        ↓
7. Validate Clean Dataset

```python
# 1. Explore Dataset
df.info()

df.describe()
```

```python
# 2. Detect Missing Values
df.isnull().sum()
```

```python
# Numerical Columns
# Replace missing values using median:
num_cols = df.select_dtypes(include=np.number).columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
# Categorical Columns
# Replace missing categories:
cat_cols = df.select_dtypes(include="object").columns

df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
# Step 3: Remove Duplicate Values

df.duplicated().sum()

df = df.drop_duplicates()
```

```python
# Step 4: Fix Outliers

Q1 = df["Order_Amount"].quantile(0.25)
Q3 = df["Order_Amount"].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
df = df[(df["Order_Amount"]>=lower) | (df["Order_Amount"]<=upper)]
```

```python
# Step 5: Fix Data Errors

# Negative quantity:
df["Quantity"] = df["Quantity"].abs()
#df["Quantity"] = df["Quantity"].apply(lambda x: x if x > 0 else -x)
#df = df[df["Quantity"]>0]

# Invalid rating:
df = df[df["Review_Rating"].between(1,5)]
```

```python
# Step 6: Fix Formatting / Data Types

# Convert date:
df["Order_Date"] = pd.to_datetime(df["Order_Date"])

# Convert category columns:
df["Customer_Gender"] = (df["Customer_Gender"].astype("category"))
```
