```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

url = "https://raw.githubusercontent.com/tirthajyoti/Machine-Learning-with-Python/master/Datasets/Mall_Customers.csv"

df = pd.read_csv(url)
df.head()
```

```python
# 1. data ingestion / acquisition

print(df.info())

print(df.describe())

print(df.shape)

print(df.isnull().sum())

print(df.duplicated().sum()) #Remove Duplicate Values

df = df.drop_duplicates()

```

```python
# 2. Data EDA

print(df.describe())

print(df.columns)

plt.figure(figsize=(8,5))
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.title("Customer Distribution")
plt.show()
```

```python
# 3. Preprocessing, if needed

X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Annual Income → similar scale
# Spending Score → similar scale
```

```python
# 4. train a model

error = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    error.append(kmeans.inertia_)


plt.figure(figsize=(12,6))
plt.plot(range(1,11), error, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Error")
plt.title("Elbow Method")
plt.show()
```

```python

kmeans = KMeans(n_clusters=5,random_state=42)
kmeans.fit(X_scaled)
```

```python
# 5. cluster the users

df['Cluster'] = kmeans.labels_

df.head()

plt.figure(figsize=(12,6))
plt.scatter(X['Annual Income (k$)'],X['Spending Score (1-100)'],c=df['Cluster'])
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("K-Means Based Customer Segmentation into Five Distinct Clusters")
plt.show()
```

```python
# Optional 6. tag each cluster with a name & save that column in dataframe

df.groupby('Cluster')[['Annual Income (k$)', 'Spending Score (1-100)']].mean()
```

```python
cluster_names = {
    0: "High Income Low Spending",
    1: "VIP Customers",
    2: "Budget Customers",
    3: "Young Potential Customers",
    4: "Average Customers"
}

# df['Cluster Name'] = df['Cluster'].map(cluster_names)

df['Customer Segment'] = df['Cluster'].map(cluster_names)

df.head()
```
