```python
import requests
import json
import pandas as pd
import time
from datetime import datetime

API_KEY = "7337ad2f4d9ba82318e5c01f3f3bead9"

BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

cities = ["New York,US","London,UK","Tokyo,JP","Sydney,AU","Paris,FR"]

weather_data = [] # to store the weather data for each city

for minute in range(15):
    print(f"Collecting weather data - Minute {minute + 1}/15")
    for city in cities:
        params = {"q": city,"appid": API_KEY,"units": "metric"} # API parameters for the request
        response = requests.get(BASE_URL,params=params) # Send the API request
        if response.status_code == 200: # Check if the request was successful
            data = response.json()
            weather_record = {
                "City": data["name"],
                "Temperature_Celsius": data["main"]["temp"],
                "Weather_Description": data["weather"][0]["description"],
                "Humidity_Percentage": data["main"]["humidity"],
                "Wind_Speed_mps": data["wind"]["speed"],
                "Timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            }
            weather_data.append(weather_record) # Append the record to the list

        else:
            print(f"Failed to retrieve data for {city}")

    if minute < 1: # original 14
        time.sleep(5)  # Sleep for 5 seconds instead of 60 for testing purposes

weather_df = pd.DataFrame(weather_data) # Create a DataFrame from the weather data

print("\nWeather Dataset:") # display the first few records of the dataset
print(weather_df.head())

print("\nDataset Shape:") # display the shape (size) of the dataset
print(weather_df.shape)

print("\nWeather Records per City:") # display the count of weather records for each city
print(weather_df["City"].value_counts())  

weather_df.to_json("weather_dataset.json",orient="records",indent=4) # 
print("\nWeather dataset successfully saved as weather_dataset.json")
```
