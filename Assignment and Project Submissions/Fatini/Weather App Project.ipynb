```python
import time
import json
import requests
import pandas as pd
from datetime import datetime

# Configuration
API_KEY = "c7b2b278d44f2483e42dc9133be0dc38"  # Replace with your actual API key
CITIES = ["New York", "London", "Tokyo", "Sydney", "Paris"]
INTERVAL_SECONDS = 60  # 1 minute
TOTAL_ITERATIONS = 15  # 15 minutes total

def fetch_weather_data():
    weather_records = []
    
    print(f"Starting weather data collection at {datetime.now().strftime('%H:%M:%S')}...")
    print("This process will take 15 minutes to complete. Please keep the terminal open.\n")
    
    for minute in range(1, TOTAL_ITERATIONS + 1):
        print(f"--- Iteration {minute}/{TOTAL_ITERATIONS} ({datetime.now().strftime('%H:%M:%S')}) ---")
        
        for city in CITIES:
            # Construct the API URL (units=metric gives Celsius automatically)
            url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={API_KEY}&units=metric"
            
            try:
                response = requests.get(url)
                
                # Check if the API call was successful
                if response.status_code == 200:
                    data = response.json()
                    
                    # Extract the requested fields
                    record = {
                        "City name": data.get("name"),
                        "Temperature (C)": data.get("main", {}).get("temp"),
                        "Weather description": data.get("weather", [{}])[0].get("description"),
                        "Humidity": data.get("main", {}).get("humidity"),
                        "Wind speed": data.get("wind", {}).get("speed"),
                        "Timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                    }
                    weather_records.append(record)
                    print(f"Successfully fetched data for {city}.")
                else:
                    print(f"Error fetching {city}: HTTP {response.status_code} - {response.text}")
                    
            except Exception as e:
                print(f"Connection failed for {city}: {e}")
        
        # Wait for 1 minute before the next loop, except after the final iteration
        if minute < TOTAL_ITERATIONS:
            print(f"Waiting {INTERVAL_SECONDS} seconds for the next interval...\n")
            time.sleep(INTERVAL_SECONDS)
            
    return weather_records

def main():
    # 1. Collect data over 15 minutes
    collected_data = fetch_weather_data()
    
    # 2. Convert the list of dictionaries into a pandas DataFrame
    df = pd.DataFrame(collected_data)
    
    # 3. Save the DataFrame directly to a JSON file
    # orient='records' saves it as a clean array of objects [{city...}, {city...}]
    output_filename = "weather_data_15min.json"
    df.to_json(output_filename, orient="records", indent=4)
    
    print("\n=============================================")
    print("DATA COLLECTION COMPLETE!")
    print(f"Total entries captured: {len(df)} (75 total rows expected)")
    print(f"Dataset successfully saved to: {output_filename}")
    print("=============================================")

if __name__ == "__main__":
    main()
```
