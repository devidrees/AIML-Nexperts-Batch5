## Natural Language Processing
- 

**Natural Language Processing (NLP)** is a subfield of AI that focuses on the interaction between *computers* and *human* (natural) languages. 

- It enables machines to **understand**, **interpret**, and **generate** human language in a way that is meaningful and useful

- NLP lies at the intersection of **linguistics**, **computer science**, and **AI**, combining the rules and structure of **human language** with **computational algorithms** to bridge the gap between human communication and machine understanding

### **Challenges**
- Issues are: 
- NL needs **context** 
    - *it is freezing here* 
        - Do you mean it is cold? temperature?
        - It is a request, please close the window

- Double meanings
    - **bat** ---> something to *play* with or bat is a *bird*
        - *"i saw a man on a hill with a telescope"*
            - you had a telescope, you used the telescope to see that man
            - there was a man on a hill, and he was carrying a telescope




- **sarcasm** 
    - you say something, but it means the complete opposite
        - oh great, a flat tire, is exactly what i needed today!!
        - did you really want that? or did you mean it is the worst thing that has happened today?
- **Tone** 
    - the attitude or emotional signal conveyed through choice of words
        - Boss: "*i would be grateful if you could provide an update*" - **polite**
            - could be said in two ways:
                - Boss asks: "*status?*" - **casual**
                - Boss says: "*as per my last email, you were supposed to report*" -  **passive aggresive**
- **Slang**
    - informal way of speaking, within a group of people
        - '*the track was on fire, no cap*' 
        - meaning: the song was excellent, i am not lying

- **Typing errors** 
    - hitting the wrong keys 
        - wanted to write i just wanted to wash my car
        - *i just wnted to wasah my cat* (r and t are close on keyboard)

- **Spelling errors** 
    - *they are going over their* (possesion ---> something they have)
    - *they are going over there* (place---> no here)

# General NLP Pipeline

Common series of steps taken for NLP. Each NLP Pipeline will be unique and will be according to the requirement:

1. Text Input - you get text from a source, document, or web pages etc

2. Text Preprocessing - text is cleaned and prepared for analysis
    - Tokenization - text is split into smaller units eg sentences, words
    - Lowercasing - every letter turned to lowercase for consistency
    - Punctuation Removal - unnecesary puncutaion is remove eg commas, period, apstrophes ,:."'-
    - Stopword Removal - common, non informative words are removed eg or, and, the  
    - Stemming/Lemmatization - words are reduced to their root forms eg. running -> run
    - Spell correction - correct the misspelled words

3. Text Representation
    - Bag of Words (BoW)
    - TF-IDF
    - Word Embeddings
    - Sentence Embeddings

4. Named Entity Recognition (NER)

5. Syntactic Analysis
    - Part-of-Speech (POS) tagging  [learn](https://www.geeksforgeeks.org/nlp/nlp-part-of-speech-default-tagging/)
    - Dependency Parsing
    - Constituency Parsing

6. Feature Engineering
    - N-gram Extraction
    - TF-IDF Scaling
    - Custom Features
    - Sentiment Scores
    - POS Tags as features
    - Text length Metrics

7. Task Specific Process
    - Sentiment analysis
    - text summarization
    - text classification
    - machine translation

8. Output generation - generate meaningful insights or predictions based on the text

## Stemming

- Reducing words to their root form by chopping off suffixes.  
- Groups similar words for analysis (e.g., *run*, *running* → *run*).  
- **Example**:  
  - Input: *"running, runner"*  
  - Output: `["run", "run"]`  

- To simplify words we cut off parts of the word, called *affixes*, to get to its root form or base word. 

**Affixes**:  
- Affixes are parts that are added to the base word (or root) to change its meaning. There are two main types of affixes:
  1. **Prefixes**: Added to the beginning of a word 
  - (e.g., "*un-*" in "*undo*").
  2. **Suffixes**: Added to the end of a word 
  - (e.g., "*-ing*" in "*running*" or "*-ed*" in "*played*").

3. **Infixes**:  
   - These are inserted *within* a word. Infixes are rare in English but are found in some other languages.  
   - Example: In some informal English, "*abso-bloody-lutely*" (inserting "bloody" for emphasis).

4. **Circumfixes**:  
   - These are added *around* a word (both at the beginning and end). Circumfixes are not common in English but can be found in other languages.  
   - Example: *en-* and *-en* are added to verbs to form the past participle (e.g., *en*light*en* from *enlighten*, meaning "to give someone greater knowledge or understanding").



![](https://i.ytimg.com/vi/B9UP8P-TAXc/maxresdefault.jpg)

**How Stemming Works**:  
- Stemming removes these affixes to get to the base or root form of the word. For example:
  - "running" → "run"
  - "happier" → "happi"  
  - "quickly" → "quick"

## Lemmatization 
- [learn](https://www.geeksforgeeks.org/machine-learning/python-lemmatization-approaches-with-examples/)

- Converting words to their base or dictionary form, considering grammar.  
- Provides contextually meaningful base forms.  
- **Example**:  
  - Input: *"better"*  
  - Output: `["good"]` 

  Lemmatization is the process of reducing a word to its base form (called a *lemma*), ensuring that the word is returned to its correct dictionary form. 
  
  Unlike stemming, which may cut off parts of words, lemmatization considers the meaning and context of the word to ensure the base form is grammatically correct.


![](https://d2mk45aasx86xg.cloudfront.net/difference_between_Stemming_and_lemmatization_11zon_393419a8d0.webp)

![](https://www.thinkdataanalytics.com/wp-content/uploads/2021/07/stemming-vs-lemmatization.png)

````python
from nltk.stem import WordNetLemmatizer

# nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

sentence = "Forests are breathing and evolving worlds that shelter countless species."
words = sentence.lower().split()

lemmatized_words = [lemmatizer.lemmatize(word, pos= 'v') for word in words]

" ".join(lemmatized_words)
````

**Word Embeddings**: 

We study methods like **Word2Vec** and **GloVe** to **represent words as vectors** in continuous space, capturing semantic relationships. These representations are crucial for understanding word meanings in context.

- `Vocabulary(V)`: This consists of all the unique words present in the corpus.
- `Corpus( C )`: All the text data or records of the dataset together are known as a corpus. e.g book
- `Document(D)`: One single text record of the dataset is a Document. e.g page or paragraph
- `Word(W)`: The words present in the vocabulary. e.g. a word

### **Why Use Word Embeddings?**
Traditional methods like one-hot encoding represent each word as a unique vector where only one element is "1" and the rest are "0". 

 ![](https://www.tensorflow.org/static/text/guide/images/one-hot.png)

This method **doesn’t capture any relationships** between words. 
For example 
- "cat" and "dog" would be equally distant from each other
- they are more similar in meaning than "cat" and "car".

With **word embeddings**, words with similar meanings **will be closer** in the vector space.
