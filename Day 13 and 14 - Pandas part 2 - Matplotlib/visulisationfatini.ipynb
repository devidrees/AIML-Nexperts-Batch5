```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("employee.csv")
df.head()

names = df["Name"]
salaries = df["Salary"]

plt.bar(names, salaries)

plt.title("Salaries of Employees")
plt.xlabel("Employee")
plt.ylabel("Salary")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")

sns.scatterplot(data = df,
                x="total_bill",
                y="tip",
                s = 50,
                hue = "day"
                )

plt.xlabel("Total Bill (RM)")
plt.ylabel("Tip (RM)")
plt.title("Total Bill vs Tip")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Select only numeric columns
corr_matrix = tips.corr(numeric_only=True)

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="plasma",
    linewidths=0.5,
    linecolor="black",
    vmin=0.5,
    vmax=1,
    cbar_kws={"shrink": 0.8},
    fmt=".2f"
)

plt.title("Correlation Matrix of Tips Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Count smokers and non-smokers
smoker_counts = tips["smoker"].value_counts()

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    smoker_counts,
    labels=["Non-Smoker", "Smoker"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Smoker vs Non-Smoker")
plt.show()
```
