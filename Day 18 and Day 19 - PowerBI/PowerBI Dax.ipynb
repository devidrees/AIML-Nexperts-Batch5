```python
// Welcome to DAX query view! Learn more about DAX queries at https://aka.ms/dax-queries.
// Right-click on tables, columns, or measures in the data pane to access quick queries, or ask Copilot for help writing DAX.

// Select "Run" to try this sample DAX query.
EVALUATE // is like SELECT
    TOPN(100, 'Global Superstore')


EVALUATE
	'Global Superstore'
	
// pd.head(5)
EVALUATE
	TOPN(5, 'superstore idrees')
	
	
// Measures - simple calculations based on columns e.g. aggregates in SQL
// MeasureName = FUNCTION('Table Name' [Column Name])


//TotalProfit = SUM('Global Superstore'[Profit])

//           table name  [measure name ] = function (table [column])
DEFINE
	MEASURE 'Global Superstore'[new2 Profit] = SUM('Global Superstore'[Profit])
EVALUATE
	{[new2 Profit]}	// { } turn a single number into a table 


// Cost measure this in DAX
DEFINE
	Measure 'Global Superstore'[Cost 3] = SUM('Global Superstore'[Sales]) - SUM('Global Superstore'[Profit]) - SUM('Global Superstore'[Shipping Cost])
EVALUATE
	{[Cost 3]}
	
	
// Selection of a few columns
EVALUATE 
	SELECTCOLUMNS(
		'Global Superstore', // gets to the table
		"First Name", 'Global Superstore'[Customer First Name],
		"product",  'Global Superstore'[Product Name],
		"Sales", 'Global Superstore'[Sales]
	)
	ORDER BY [Sales] 
	// order by a measure -- table name[measure]
	// order by a column -- [column]
	
	
// Order by on a table	
EVALUATE
 'Global Superstore'
ORDER BY 
	'Global Superstore'[Sales] DESC
	

// GROUP BY - 
EVALUATE 
	FILTER(
	SUMMARIZE(
		'Global Superstore', // gets to the table
		'Global Superstore'[Category], // Column
		"Sales", SUM('Global Superstore'[Sales]), // name aggregate , FUNCTION(Column)
		"Cost", SUM('Global Superstore'[Cost])
		),
		[Cost] > 3385360
	)
		ORDER BY [Cost]
		

// FILTER 
EVALUATE 
	FILTER(
		TOPN(50, 'Global Superstore'), // table part
		'Global Superstore'[Cost] < 500 // filter expression
	)
	ORDER BY
	[Cost] ASC
	


	
```
