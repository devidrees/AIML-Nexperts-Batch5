- First you check for inconsistencies in data
    - Missing Values
    - Duplicates
    - Erros in your data
    - Outliers

- Every dataset has its identity:
    - Data Distribution 
        - how is the data spead about - what is the std deviation, variance
    - Central Tendencies
        - where the centre of the data is located
        - mean median mode --- represent the centre of the data
    - Data Transformations: (Feature Scaling is one of things we do here)
        - [Standardization vs Normalization](https://www.geeksforgeeks.org/machine-learning/normalization-vs-standardization/) 
        - Normalization - Adjust values to a common scale
            - min max scaling
            - entire data is squished between 0 and 1 
            - This scales the range to [0, 1] or sometimes [-1, 1]
        - Standardization - Convert data to a standard form for a common scale
            - mean 0 and std deviation 1
            - 

- Feature Scaling - take a feature/column and you transform it entirely

### Handling Missing Values
- Deletion (Removal of these missing values)
    - Listwise:
        - Rows where any data is missing
    - Pairwise:
        - remove specific fields with missing values, keep the rest
- Imputation  (filling in the missing values)
    - estimate/calculate/predict the values that should have been there
    - should not affect the identity of the data set too drastically
    - distibution, the central tendencies should remain almost intact
    - Types:
        - Mean/median/mode imputation - put these values in missing places
        - Predective imputation - using ML we predict the missing values first
        - Forward/Backward fill - propagage the values - fill in with the previous or the next one
