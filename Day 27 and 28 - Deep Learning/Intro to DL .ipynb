| **Hardware Dependency**  | Can run on traditional hardware setups.                   | Requires specialized hardware like GPUs/TPUs.         |
| **Performance**          | Good for smaller datasets and simple tasks.              | Excels with large, complex datasets and unstructured data. |

We wanted to model this neuron. To mimic how humans think.

- Frank Rosenblatt (1928 – 1971) was an American psychologist notable in the field of Artificial Intelligence.
- 
- In 1957 he started something really big. He "invented" a Perceptron program, on an IBM 704 computer at Cornell Aeronautical Laboratory
- 
- Scientists had discovered that brain cells (Neurons) receive **input** from our **senses** by electrical signals.
- 
- use electrical signals to store information, and to make decisions based on previous input.

Frank Rosenblatt suggested this algorithm:

1. Set a threshold value  ---> Decision Boundary
2. Multiply all inputs with its weights ---> Dot Product
3. Sum all the results ---> Weighted Sum - you add according to the weightage/importance
4. Activate the output ---> whether the threshold is met or not
- This was called a Perceptron - an **Artificial Neuron**
    - The smallest unit of a Neural Network modeled on the human brain cell

- A Neural Network in its simplest form - an Artificial Neural Network (ANN) unit

1. Set a threshold value
2. Multiply all inputs with its weights
3. Sum all the results
4. Activate the output
- **Make a decision, weather you want to go to a concert**


Criteria	| Input (binary) |	Weight (out of 1)
-----|----|-----
Artists is Good	| x1 = 0 or 1 |	w1 = 0.7
Weather is Good	| x2 = 0 or 1	| w2 = 0.5
Friend will Come |	x3 = 0 or 1	 | w3 = 0.6
Food is Served	| x4 = 0 or 1	 | w4 = 0.3
Juice is Served |	x5 = 0 or 1	 | w5 = 0.4

```python
import numpy as np

inputs =  np.array([1, 1, 1, 0, 0]) # depends on situation # can be anything
weights = np.array([0.7, 0.5, 0.6, 0.3, 0.4 ]) # we assign weights ourselves

# let's apply the frankrosenblat algo 
# Let's use a Perceptron to make our decision

# Step 1: Threshold
threshold = 1.5

# Step 2: Multiply inputs with weights
product = inputs * weights # broadcasting # [0.7 , 0, 0.6, 0, 0]

# Step 3: Weighted sum
weighted_sum = sum(product)

# Step 4: activate the perceptron
def step_activation(x):
    if x >= threshold:
        return "Yes, We will go to the Concert!"
    else:
        return "No. We ain't going."

# apply the activation
step_activation(weighted_sum)
```

```python
# another implementation
# combined in one function

def perceptron(artist, weather, friend, food, juice, threshold = 1.5):
    inputs = np.array([artist, weather, friend, food, juice])
    # the neuron already has the importances
    weights = np.array([0.7, 0.3, 0.8, 0.3, 0.4])

    # step2: multiply the inputs and weights
    product = inputs * weights

    # Step3: weighted sum
    weighted_sum = sum(product)

    # step4: activate the output
    if weighted_sum >= 1.5:
        return "Yes, we go!!!"
    else: 
        return "NOPE, we are not going!"


# whether is good
decision_sunny = perceptron(artist = 0, weather = 1, friend = 1, food = 1, juice = 1)
decision_sunny
```

### Different types of Neural Networks
- We can call them DNN (Deep Neural Networks) if they have more than atleast 4 layers - can be 20 to 50 to 100 layers as well, depending on data volume
- 
1. *Feedforward* Neural Network / Multi Layer Perceptron
    - simplest form of a neural network
    - data or input goes from the start through the layers to the output of the model
    - although MLPs can be given a *backpropagation* algorithm as well
    - 

2. Convolutional Neural Networks (CNN)
    - that works best on grid like data
    - it automatically extracts important features from data
    - e.g. images

3. Recurrent Neural Network (RNN)
    - with loops to allow retention of information
    - Simple RNN - used for simple patterns and short sequences
    - LSTM - used for long sequences, complex patterns
    - GRU - used for speech recognition
    - Bidirectional RNN - context/ from both sides is retained
    - e.g. Time series data, textual data

4. Transformers
    - networks based on attention mechanisms
    - learn relationships between elements parallelly 
    - use parallel processing, no recrurrence
    - e.g. Transfer learning - pretrained models 
        - Pretrain models once, then use your own data to finetune
    - NLP tasks: text classification - BERT, GPT
    - Vision tasks: Image classification - ViT

5. Autoencoders
    - trained to reconstruct their input
    - e.g. noise removal, generative modeling (generate images or data)    

6. Generative Adversarial Networks (GAN)
    - two networks Generator and Discrimnator
    - one generated data and another judges it
    - e.g. Vanilla GAN - simple image generation
        - DCGAN - higher quality
        - CycleGAN - image to image translation or style transfer
