#### common files in Projects or Libraries

`__init__.py` ---> Dunder Files
- the Receptionist 
- existance of this file, tells python that this is an official packages
- i.e. its a library we can use
- this script automatically runs when we import the package
- 
`__main__.py`
- the Command Center of the package
- This is like a switch you have to turn ON
- you execute this file, to start something
    - example:
        - a multi agent system - has a lot of programs
        - you want to execute some specific functions
        - you save those in the main, and execute the main
        - main will execute all those functions automatically
- main.py is a script
    - how do we execute scripts?
    - go to terminal
    - go to the folder where the script exists
    - write: `python <script name.py> `
    - another is to download a RUNNER Extension in VS Code

- In any module file, if there is a function, it will be executed when you import the module(script)
    - what shall we do
    - whatever program there is to run in a script, we will keep it in a code block of this:
        - `if __name__ == "__main__":`
