````python
class Possesion:
    # Class attribute
    # this variable will be shared among all the methods
    num_possesions = 0 # Global or Shared Attribute

    # python's way of understanding, what you want to take into the function -- or what inputs/variables
    # __init__ : it is a Constructor --- helps us build objects later on
    # everytime we create an object, this function is executed
    def __init__(self, kind, age, color, brand): # this is the place to put any parameters
        # self here targets the object that we will create later
        # save the data for an object
        self.kind = kind # attributes of the class
        self.age = age
        self.color = f"this is "
        self.brand = brand

        # Possesion.num_possesions = Possesion.num_possesions + 1
        Possesion.num_possesions += 1
        print(f"A {self.kind} was added")

    # variable inside a method --- we are not using any attributes from objects here
    def greet():
        lucky_number = 42 # local to this method
        print(f"{lucky_number} says Hello!!! and Possesions are currently {Possesion.num_possesions}")

    # access attributs of an object inside a method
    # we are using attributes from objects here
    def details(self):
        # this should give most important details of an object
        # new_number = 23
        print(f"Type is: {self.kind}\nThe age is: {self.age}\nThe brand is: {self.brand}")
    
    # Getter setter methods! 
````

### Inheritance and Polymorphism
Inheritance
- a new class (CHILD) can reuse code from a existing class (PARENT)
- 
Polymorphism
- the same method, gives different outputs

##### Inheritance
```python
super().__init__()  # is used to call a parent class
```
- example:
- Parent===== Animal
- Children Classes === Dog, Cat etc

````python
# Parent class
class Animal:
    def __init__(self, name, species, age):
        self.name = name
        self.species = species
        self._age = age 

    def introduce(self):
        print(f"Hi. I am {self.name}!!\nI am a {self.species}\nI am {self._age} years old!")

    # Getter Method
    def get_age(self):
        return self._age
    
    # Setter Method
    def set_age(self, new_age):
        self._age = new_age
        return f"The new age is {new_age}"
    
    def have_birthday(self):
        self._age +=1
        print(f"It's {self.name}'s birthday. Now {self._age} years old ")
````

````python
# CHILD of ANIMAL
class Dog(Animal):

    def __init__(self, name, age): # this is what we will pass ourselves
        # super() refers to the parent class --- could have used Animal as well
        # 
        self._age = age
        super().__init__(name, "Dog", age)

    def sound(self):
        print(f"{self.name} says: Woof!!")

    # different methods for different classes
    def detect_drugs(self):
        if self._age > 5:
            return f"I will sniff out any drugs"
        else:
            return f"{self.name} says, I am just {self.age} years old. You fr??"
````

````python
# CHILD of ANIMAL
class Cat(Animal):

    def __init__(self, name, age):
        super().__init__(name,"Cat", age)

    def sound(self):
        print(f"{self.name} says: Meowww!!")
````

````python
# CHILD of ANIMAL
class Parrot(Animal):

    def __init__(self, name, age):
        super().__init__(name,"Parrot", age)

    def sound(self):
        print(f"{self.name} says: Mithu Kakaaa!!")
````

````python
# examples of working with a class
dog1 = Animal("Bruno", "Dog", 5)
````

````python
# this goes against the concept of ENCAPSULATION
# breaks ENCAPSULATION
dog1._age = 6
dog1._age
````

````python
dog1.get_age()
dog1.set_age(8)
dog1.introduce()
````
