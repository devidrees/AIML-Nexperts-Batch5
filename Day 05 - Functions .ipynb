````python
# Create a function that gives the sum of n numbers
# Sum of first n numbers ----> (n* (n +1)) // 2
# sum_n(10) -> 1 + 2 + 3 + ...+ 10 = 

def sum_n(n):
  sum = 0
  for i in range(n+1):
    sum = sum + i
  return sum

sum_n(100_000_000)
````

````python
# Prime number function
# We pass a number, and the function says if its prime or not
 
def prime(num):
    if num == 1 or num == 2 or num == 3:
        print('Prime Number')
    elif num % 2 == 0 or num % 3 == 0:
        print('Non Prime Number')
    elif num % 1 == 0 and num % num == 0:
        print ('Prime Number')

prime(23)

# check all the prime numbers upto 30
# 2 Prime
# 3 Prime
# 4 Non Prime
````

````python
def isPrimeNumber(num):
  if num > 1:
    for i in range(2,num):
      if (num % i) == 0:
        print(num,"is not a prime number")
        print(i,"times",num//i,"is",num)
        break
    else:
      print(num,"is a prime number")
  else:
    print(num,"is not a prime number")
 
for i in range(31):
  isPrimeNumber(i)
````
