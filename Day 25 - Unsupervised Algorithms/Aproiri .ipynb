```python
import pandas as  pd

# encode data in a form Model can understand
from mlxtend.preprocessing import TransactionEncoder

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# dataset of transactions

transactions = [
    ['Milk', 'Butter','Bread'],
    ['Milk','Bread'],
    ['Bread', 'Butter', 'Eggs'],
    ['Milk', 'Eggs'],
    ['Bread', 'Butter', 'Juice']
]
```

```python
# EDA -- 
# Data Cleaning --- > Data transformation
# Algorithms cannot read words - we convert the data into a True False 0 1 table 
te = TransactionEncoder()

te_array = te.fit_transform(transactions) #.transform(transactions) 

# create a dataframe of these encoded transactions along with table header
df_transactions = pd.DataFrame(te_array, columns = te.columns_)
display(df_transactions)
```

```python
# Find frequently bought items

# Run the Apriori
frequent_itemsets = apriori(df_transactions, 
                            min_support = 0.4, # we only care about the items bought in 40 % of transaction
                            use_colnames = True)
```

```python
clean_itemsets = frequent_itemsets['itemsets']
#  clean the frozenset output
```

- Support: How often the rule appears in the dataset.
- Confidence: Probability of buying item B if item A is bought.
- Lift: Strength of the rule over random chance. (>1 means it's a good rule)

```python
# generate association rules
# rules to find what is likely going to be bought
rules = association_rules(frequent_itemsets, # dataframe of itemsets and support
                          metric = 'confidence',
                          min_threshold = 0.4, # atleast 70 % confidence
                          )

final_table = rules[['antecedents',	'consequents',	'support',	'confidence','lift']]

display(final_table)
```

```python
# test our system

current_cart = {'Bread'} # frozenset({'Bread})

# find consequent based on the antecedant
recommendations = final_table[rules['antecedents'] == current_cart]

recommendations
```
