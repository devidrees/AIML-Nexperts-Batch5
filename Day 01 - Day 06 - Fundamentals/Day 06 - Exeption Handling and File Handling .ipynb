## Exception Handling

When an error occurs, or exception as we call it, Python will normally stop and generate an error message.


```python
try:
    #code
except SomeError:
    # what to do or error msg

- Common examples of errors 
1. Dividing by Zero - ZeroDivisionError
2. Converting str to int - TypeError
3. Opening file that does not exist - FileNotFoundError
4. Code that is not valid - SyntaxError

````python
try:
    x = int(input("enter a number"))
    y = 10 / x
    print("Result = ", y)
except:
    print("Something is Wrong!!!")
````

- Handling errors delibrately 
- we don't let the program bomb or crash
- Crashing gracefully

````python
try:
    x = int(input("enter a number"))
    y = 10 / x
    print("Result = ", y)
    if x == 1:
        print('yaay zero!!')

# except ZeroDivisionError:
#     print("you divided with 0, you dummy!!!")
except TypeError:
    print("you divided a number with a alphabet or string, you dummy!!!")
except Exception as e:
    print("You don't know how to code. go train yourself first!!", e)
````

- Most common and important use of exception handling you will see in:
    - say your program depends on a website and that website is down
        - you keep that line of code in a try except block
    - if your program depends on a file
        - csv files - contain - what if it does not exist
        - the line of code accssing this file should be kept in try except as well.

## File Handling:
- how does python access a file
- operations like:
    - READ
    - WRITE
    - OPEN
    - CLOSE

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

````python
try:
    file = open("idrees.txt", "r") 
except Exception as e:
    print("Something is wrong.", e)
````

````python
# CREATE - a file
file = open("idrees.txt","w") # write - "w"
````

````python
# OPEN - a file
file = open("idrees.txt", "r")  # read - "r"
````

````python
# WRITE - on a file
# open in write mode
file = open("idrees.txt", "w+") # open in write + read mode

file.write("Idrees is an awesome TEACHERRRRRRR\n")
````

Context Manager - using the `with`
- with open( <filename> , "r+") as file:
        file.read()
        file.write()
        file.close # is done automatically and you don't need to close files

- automatically closes

````python
with open("idrees.txt", "w+") as f:
    f.write("Idrees is awesome\n")
    f.write("idrees wins all the battles he fights\n")
````

````python
with open("idrees.txt", "r") as f:
    content = f.read()
    print(content)
````

````python
with open("idrees.csv", "w+") as f:
    f.write("name, age, subject\n")
    f.write("idrees, 30, AIML\n")
````
