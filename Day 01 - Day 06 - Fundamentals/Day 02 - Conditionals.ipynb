````python
# if else ladder

age = int(input("enter your age"))

if (age <= 19):
  print("You can a Teen")
elif (age < 40):
  print("You are a young adult")
elif (age < 75):
  print("You are an adult")
elif (age < 85):
  print("You are a senior citizen")
else:
  print("You are Old")
````

# Comparison Operations
- Values we compare ---> bool we get
- always return a bool value
- Operators:
<,

<=,

>,

>=,

 == (is equal to),

 != (is not equal to)

````python
10 <= 12

10 >= 12

10 == 12

10 != 10
````

````python
# can we compare Bools?

True  == False

True  == True

True == True

True > False  # 1 > 0

True < False

False > True
````

````python
# REMEBER: ALPHABETIC Order decides what is greater

'a' < 'b'
'z' > 'a'
# whatever comes later has more value in value - for comparison operation

'a' < 'z'

ord('A') # Gives you the ASCII code for that character

'apple' > 'ball'

'apple' > 'aapple'
````
