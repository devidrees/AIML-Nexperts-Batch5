### Loop Control
- break - stop the loop early
- continue - skips to the next iteration
- pass - does nothing, just helps execute instead of error

````python
for i in range(0,10,2): # range(start, end, stepsize)
    pass
````

````python
for i in range(0,10):
    print(i)
    if i == 5:
        break
````

````python
for i in range(0,10):
    if i == 5:
        break
    print(i)
````

````python
for i in range(0,10):
    if (i == 5) or (i == 3) or (i == 4) :
        print("Hello ", i)
        continue
    print(i) # this statement is skipped
````

````python
i = 0
while i < 10:
    i += 1 # same as i = i + 1
    if (i == 5) or (i == 3) or (i == 4) :
        print("Hello ", i)
        continue
    print(i) # this statement is skipped
    if i == 7:
        break
````

- Input validation - to make sure the correct input goes into a program
    - only the correct input goes through
- Execption Handling - targets the error that can come up

- when writing code for a strong program, or a good software, these two methods are very important

````python
# how to use break, continue, input validation, and exception handling
while True:
    num = input("enter a number") # string

    # check for exit
    if (num == "quit") or (num == 'q'):
        print("Program Stopped!")
        break

    # validate if input is an integer
    # consider all wrong things users can send
    # alphabets or words or symbols or numbers
    if not num.isdigit():
        print("Not an INTEGER. Try again.")
        continue

    # Exception Handling - if we did not have input validation
    try:
        print("Square is: ", int(num) **2)
        print("Cube is: ", int(num) **3)
        print("Square Root is: ", int(num) **0.5)
    except ValueError:
        print("That is not a valid Input. Enter a number.")
````

5. FizzBizz program: Ask user for input.
    - Numbers range from 0 to 20
    - if input is divisible by 3, print 'Fizz'
    - if divisible by 5, print 'Buzz'
    - if divisible by 3 and 5 both, print 'FizzBuzz'

````python
number = int(input("Enter a number between 0 to 20: "))
word = ""

if (number % 3) == 0 and (number % 5) == 0:
  word = "FizzBuzz"
elif (number % 5) == 0:
  word = "Buzz"
elif (number % 3) == 0:
  word = "Fizz"
else:
  word = "Invalid"
 
print(word)
````

````python
while True:
 
    number = int(input('Enter a number'))

    if number == -1:
         break
 
    try:
        if number == 0:
            pass
        elif number % 3 == 0 and number % 5 == 0:
            print('FizzBuzz')
        elif number % 5 == 0:
            print('Buzz')
        elif number % 3 == 0:
            print('Fizz')
        else:
            print("Enter a new number")
    except ValueError:
            print('Please enter a valid number')
            continue
````
