## Keyword Arguments and Default Arguments
- *kwargs
- you give a name to the argument
- 

````python
def summing(num1, num2, power): # parameters
    sum = (num1 + num2) ** power
    return sum


# function call
summing(1, -2, 3) # order is important
# simple arguments

# keyword arguments --- Order is immaterial/not important
summing(power=3, num1= -3, num2= 1) # but there is a keyword to them as well
summing(1,2,3)

# ------------------------------------------
# with default arguments
def summing(num1 = 1, num2 = 1, power = 1): # parameters
    sum = (num1 + num2) ** power
    return sum


summing(3,2,2)
````

````python
# Prime number function
# We pass a number, and the function says if its prime or not
 
def prime(num):
    if num == 1 or num == 2 or num == 3:
        print('Prime Number')
    elif num % 2 == 0 or num % 3 == 0:
        print('Non Prime Number')
    elif num % 1 == 0 and num % num == 0:
        print ('Prime Number')

prime(23)

# check all the prime numbers upto 30
# 2 Prime
# 3 Prime
# 4 Non Prime
````

Map() - 
- map() - is used for transforming - Applies a **function** to an **iterable**
- map(function, iterable) # map(square, num_list)

Filter() - 
- filter() - used for removing - Applies a rule/ expression to an iterable

````python
numbers = [1, 2, 3, 4, 5]

def double(num):
    return num *2

double(5)

list(map(double, numbers))
````

````python
list(map(lambda x: x * 2, numbers))
````

````python
numbers = [1, 2, 3, 4, 5]

def odds(num):
    if num % 2 != 0:
        return num

odds(4)

list(map(odds, numbers))

list(filter(odds, numbers))
````

````python
odds = list(filter(lambda x: x % 2 != 0, numbers))
odds
````

## Function Recursion

- function calls itself
- Remember:
    - Base Case: when to stop
    - Recursive case: do a small part, and call function

```python
def function(<parameters>):
    if <condition>:
        return result
    else: 
        return function(<modified parameters>)

https://www.w3schools.com/python/python_recursion.asp

````python
# 10 -> 0
def countdown(n):
    print(n)
    # base case - to exit the infinite loop
    if n == 0:
        print("DONE!!!")
        return n

    # recursive case - with updated parameters
    countdown(n - 1) # parameters are updated
````

````python
countdown(10)
````

````python
n = 5
factorial = 1
for i in range(1, n+1):
    factorial = factorial * i

print(factorial)
````

````python
# factorial using for loop
def multiplyn(n):
    factorial = 1
    for i in range(1, n+1):
        factorial = factorial * i
    return factorial

multiplyn(10)
````

````python
multiplyn(20)
````

````python
# factorial using for loop
def multiplyn(n: int):
    # base case
    if n == 1:
        return n
    
    # recursive
    return n * multiplyn(n - 1)

multiplyn(5.0)
````
