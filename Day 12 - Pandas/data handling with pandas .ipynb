## Selection using the Location or Labels
##### integer Location --- **.iloc[]**
- `df.iloc[row index]` # access rows
- 
- `df.iloc[row_index_slice]` # [start: end : stepsize]
- INDEXING ----
- `df.iloc[row_slice, column_slice]`
- 

```python
df.iloc[0] # first row as a series object

df.iloc[-1] # last row

df.iloc[1:3] # rows from 1 upto 3

df.iloc[1:4:2] # in steps of 2

# ---- indexing works

df.iloc[:4, 1:3] #first 4 rows, 

df.iloc[[0,2,3], [0,2,3]] # you can access the rows, columns by number, but
# if out of index, error will be thrown
# errors are not thrown if you are accessing using slices
```

```python
# i want the data of the last two entries, for department and experience
df.iloc[-2:,[1,3]]
```

#####  Label - .loc[]
- focuses on labels
- df.loc[0:3] # targets the names of these rows
- Which are the indexes
- 

```python
# 
ind = ['a','b','c','d','e']
df = pd.read_csv("employee.csv")
df['index'] = ind

df = df.set_index('index')
display(df)
# -----
df.index = ['a','b','c','d','e']
display(df)
```

```python
df.loc[0] # row
df.loc[0:3] # slice of rows

# df.loc[0:3, 0:3] # throws an error

df.loc[0:3, ['Department', 'Experience']]
```

```python
df.loc['a':'d', 'Department':'Experience':2]
```
