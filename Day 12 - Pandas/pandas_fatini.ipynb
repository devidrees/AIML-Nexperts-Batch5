```python
import pandas as pd

maths = pd.Series([20, 30, 40], dtype='int8', name = 'Maths')
type(maths)

print(maths)
```

```python
english = pd.Series([20,30,40])
python = pd.Series([20,30,40])
science = pd.Series([20,30,40])
```

```python
pd.concat([maths, english, python, science], axis=1)
```
